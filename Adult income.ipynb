### Replace income values ["<=50K" with 0][">=50K" with 1]

```python
df1["income"].unique()
df1["income"].value_counts()
```

```python
ax1 = sns.countplot(x = "income", data=df1)

for bars in ax1.containers:
    ax1.bar_label(bars)
```

```python
# method 1
##df1["income"].replace("<=50K", 0)
##df1["income"].replace(">=50K", 0)

##OR

#df1["income_fact"] = df1["income"].replace(to_replace=["<=50K", ">50K"], value=[0,1])
```

```python
# method 2 
def income_data(x):
    if x =="<=50K":
        return 0
    else:
        return 1
    
df1["income"] = df1["income"].apply(income_data)

#df1["income_category"] = df1["income"].apply(income_data) #if you want to add it as a new column
df1
```
